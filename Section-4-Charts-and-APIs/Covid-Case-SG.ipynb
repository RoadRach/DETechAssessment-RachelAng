```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import datetime as datetime
import json

url = 'https://api.covid19api.com/country/singapore/status/confirmed'

resp = requests.get(url)
jdata = resp.json()
```

```python
df = pd.json_normalize(jdata)
# timeLine = datetime.datetime.year(df['Date'])
# df.info()
# df.head()
print(df)

X1=df['Date'].str[:4]
Y=df['Cases']

fig, ax = plt.subplots()
ax.plot(X1, Y, color='green')

plt.ylabel("Number of Cases")
plt.xlabel("Year")
fig.savefig("Covid_Singapore.png")
fig.show()

```

```python
fig2, ax2 = plt.subplots()
X2=df['Date'].str[:7]
fig2 = ax2.scatter(X2, Y)
plt.ylabel("Number of Cases")
plt.xlabel("Year-Month")
plt.figure(figsize=(10,5))
plt.show()
```
