```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests,datetime,os

url = 'https://api.covid19api.com/country/singapore/status/confirmed'

resp = requests.get(url)
jdata = resp.json()
```

```python
df = pd.json_normalize(jdata)
# timeLine = datetime.datetime.year(df['Date'])
df.info()

X=df['Date'].str[:7]
Y=df['Cases']

fig, ax = plt.subplots()
ax.plot(X, Y, color='green')
fig.savefig("Covid_Singapore.png")
fig.show()
```
